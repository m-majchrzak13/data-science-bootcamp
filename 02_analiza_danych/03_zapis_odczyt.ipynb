```python
import numpy as np
import pandas as pd
```

```python

def fetch_financial_data(company='AMZN'):
    """
    This function fetch stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('META')
df.head()
```

```python
df.to_csv('meta.csv')
```

```python
df_aug = df[(df.index.month == 8) & (df.index.year == 2025)]
df_aug
```

```python
df_aug.to_csv('meta_aug.csv')
```

```python
new_df = pd.read_csv('meta_aug.csv', index_col=0)
new_df
```
